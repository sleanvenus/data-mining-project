```python
import pandas as pd

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
onehot_encoded_train = pd.get_dummies(df_train)
onehot_encoded_test = pd.get_dummies(df_test)
```

```python
import lightgbm as lgb

label = onehot_encoded_train['exceeds50K']
data = onehot_encoded_train.drop(['exceeds50K'], axis=1)

data, onehot_encoded_test = data.align(onehot_encoded_test, join='inner', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, label, test_size=0.2)
```

```python
d_train = lgb.Dataset(X_train, label=y_train)
d_test = lgb.Dataset(X_test, label=y_test)

parameters = {
    'application': 'binary',
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 30,
    'feature_fraction': 0.5,
    'bagging_fraction': 0.5,
    'bagging_freq': 20,
    'learning_rate': 0.05,
    'verbose': 0
}
clf = lgb.train(parameters,
                        d_train,
                        valid_sets=d_test,
                        num_boost_round=5000,
                        early_stopping_rounds=380)
```

```python
import numpy as np

ids = list(range(1, df_test.shape[0] + 1))
y = clf.predict(onehot_encoded_test)
y = np.where(y > 0.5, 1, 0)
output = pd.DataFrame({'id': ids, 'prediction': y})
output.to_csv("submission.csv", index=False)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
validation_y = np.where(clf.predict(X_test) > 0.5, 1, 0)
classification_report(y_test, validation_y, digits=4)
```

```python
confusion_matrix(y_test, validation_y)
```

```python
accuracy_score(y_test, validation_y)
```
